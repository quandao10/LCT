```python
import torch
import matplotlib.pyplot as plt

# Define parameters for the underlying normal distribution
mu = 0.0     # Mean of the log of t
sigma = 0.25  # Standard deviation of the log of t

# Create a LogNormal distribution object in PyTorch
lognormal = torch.distributions.LogNormal(mu, sigma)

# Number of samples to generate
n_samples = 1000

# Sample values from the lognormal distribution
samples = lognormal.sample((n_samples,))

# Display the first few samples
print("First 10 samples:", samples[:100])

# Plot a histogram of the samples
plt.figure(figsize=(10, 6))
plt.hist(samples.numpy(), bins=50, edgecolor='black', alpha=0.7)
plt.title('Histogram of Samples from a Lognormal Distribution')
plt.xlabel('Sample value')
plt.ylabel('Frequency')
plt.show()
```
